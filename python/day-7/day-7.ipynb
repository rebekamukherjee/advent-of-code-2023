### --- Day 7: Camel Cards ---

https://adventofcode.com/2023/day/7

```python
def get_hand_type(hand):
    all_cards = list(hand)
    unique_cards = set(all_cards)
    if len(unique_cards)==1:
        return "five of a kind"
    elif len(unique_cards)==2:
        for card in unique_cards:
            if all_cards.count(card)==4:
                return "four of a kind"
        return "full house"
    elif len(unique_cards)==3:
        for card in unique_cards:
            if all_cards.count(card)==3:
                return "three of a kind"
        return "two pair"
    elif len(unique_cards)==4:
        return "one pair"
    elif len(unique_cards)==5:
        return "high card"
    
hand_to_bid_dict = {}
input_text = open('input.txt', 'r').read()
input_text = input_text.split('\n')
for line in input_text:
    hand, bid = line.split()
    if hand in hand_to_bid_dict.keys():
        print (hand)
    hand_to_bid_dict[hand] = int(bid)
type_to_hand_dict = {
    "five of a kind": [],
    "four of a kind": [],
    "full house": [],
    "three of a kind": [],
    "two pair": [],
    "one pair": [],
    "high card": []
}
for hand in hand_to_bid_dict.keys():
    hand_type = get_hand_type(hand)
    type_to_hand_dict[hand_type].append(hand)
weakest_to_strongest_hands = [
    "high card",
    "one pair",
    "two pair",
    "three of a kind",
    "full house",
    "four of a kind",
    "five of a kind"
]
card_sort_order = {'2':0, '3':1, '4':2, '5':3, '6':4, '7':5, '8':6, '9':7, 'T':8, 'J':9, 'Q':10, 'K':11, 'A':12}
def custom_card_sort_key(item):
    return [card_sort_order[char] for char in item]
total_winnings = 0
new_hand_type_start_rank = 1
for hand_type in weakest_to_strongest_hands:
    hands = type_to_hand_dict[hand_type]
    sorted_hands = sorted(hands, key=custom_card_sort_key)
    for rank in range(len(sorted_hands)):
        hand = sorted_hands[rank]
        bid = hand_to_bid_dict[hand]
        true_rank = new_hand_type_start_rank+rank
        winnings = true_rank * bid
        total_winnings += winnings
    new_hand_type_start_rank+=len(sorted_hands)
print("The total winnings = ", total_winnings)
```
